```python

dbox_link='https://www.dropbox.com/scl/fo/xpsvbsuu6nzsl4rtm7fvv/h?rlkey=wj4gdwxalt4wse40um2z2fvjx&dl=1'

def dbox_to_path(dbox_link,path):
    import urllib.request
    zipped_file_path = path + ".zip"
    urllib.request.urlretrieve(dbox_link, zipped_file_path)
    !echo A | unzip $zipped_file_path -d $path
    return path



def name_corrections(dir_link,str_rep_dict={}):
    import os
    import google
    path = dbox_to_path(dir_link,'my_files')
    for k,v in str_rep_dict.items():
        for d,r,files in os.walk(path):
            for f in files:
                if k in f:
                    new_name = f.replace(k,v)
                    os.rename(os.path.join(d,f),os.path.join(d,new_name))
    !zip -r /content/renamed.zip /content/my_files
    google.colab.files.download("/content/renamed.zip")



str_rep_dict={'WT_Sag_P586':"WT_F_P586_Sag",
              'WT_Sag_P590_F':'WT_F_P590_Sag',
              'WT_Sag_P590_s':'WT_F_P590_Sag_s'

              }
name_corrections(dbox_link,str_rep_dict=str_rep_dict)
```
